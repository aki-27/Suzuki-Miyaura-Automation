```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('192experiments_average.csv').dropna()
df
```

```python
df2 = df.to_numpy()
x = list( df2[:,0] )
y = list( df2[:,1] )
```

```python
plt.rcParams['font.family'] = 'arial' # font family
plt.rcParams['xtick.direction'] = 'out' #x
plt.rcParams['ytick.direction'] = 'out' #y
plt.rcParams['font.size'] = '8.0'

fig, ax = plt.subplots( dpi=300, figsize=(7,3.5) )
ax.plot( x, y, color = 'coral', zorder=2 )
ax.scatter( x, y, color = 'dodgerblue', zorder=3, s=8 )
labels = [1, 8, 16, 24, 32, 40, 48]

ax.set_xticks( labels, map( str, labels ) )
ax.set_ylim( ymax = 40 )
ax.set_xlabel( 'Number of experimental batches' )
ax.set_ylabel( 'Average of 4 observed yields of $\mathbf{3}$' )

ax.grid( axis='both', zorder=1, color='silver', linewidth = 0.4 )

plt.savefig( 'optimization_profile_average_of_4samples.png', bbox_inches='tight' )
```
