# 疑似データの作成

```python
import pandas as pd

df = pd.read_csv('2024_0712_0146_candidates.csv')
```

```python
df_res = df.dropna()
df_res
```

```python
mask = pd.isnull( df['objectives'] )
df_blank = df[mask]
df_blank
```

```python
df_res.iloc[:,4:12]
```

```python
import physbo
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

X = df_res.iloc[:,4:12].to_numpy()
y = df_res.iloc[:,12].to_numpy()

cov  = physbo.gp.cov.gauss( X.shape[1], ard = False )
mean = physbo.gp.mean.const()
lik  = physbo.gp.lik.gauss()

gp = physbo.gp.model( lik=lik, mean=mean, cov=cov )
config = physbo.misc.set_config()


gp.fit( X, y, config )
gp.prepare( X, y )
y_pred = gp.get_post_fmean( X, X )

from sklearn.metrics import r2_score
print( 'R(train all):', r2_score(y, y_pred) )

plt.xlabel("expt")
plt.ylabel("predict")
plt.scatter( y, y_pred )
```

```python
X_blank = df_blank.iloc[:,4:12].to_numpy()
y_blank = gp.get_post_fmean( X, X_blank )
y_blank
```

```python
df_blank.to_csv('out.csv')
pd.DataFrame( y_blank ).to_csv('out_y.csv')
```

# 予測のループドライラン

```python
import physbo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def load_data():
    A = pd.read_csv('completed_only_num.csv').to_numpy()
    X = A[:,4:12]
    t = A[:,12]
    #A = np.asarray(np.loadtxt('s5-210.txt', skiprows=1, delimiter=','))
    #X = A[:,0:3]
    #t = A[:,3]
    
    return X, t
```

```python
class simulator:
    def __init__(self):
        _, self.t = load_data()
    def __call__(self, action):
        return self.t[action]
        
```

```python
for z in [5,4,3,2,1]:
    X, t = load_data()
    fig, ax = plt.subplots()
    res_all = []
    
    total_size = 512
    batch_size = 2**z
    loop_num = 30
    run_num = int(total_size / batch_size)

    print( 'loop  =', loop_num )
    print( 'batch =', batch_size )
    print( 'run num =', run_num )
    
    from mpi4py import MPI
    
    df_objects = pd.DataFrame()
    df_actions = pd.DataFrame()
    df_best_objs = pd.DataFrame()
    df_best_acts = pd.DataFrame()
    score_mode = 'EI'

    print( '===== BATCH SIZE ' + str(batch_size) + ' =====' )
    
    for i in range( loop_num ):
        print( '----- SAMPLE ' + str(i+1) + ' ------' )
        
        X_scaled = physbo.misc.centering( X )
        policy = physbo.search.discrete.policy( test_X = X_scaled, comm=MPI.COMM_WORLD )
        policy.set_seed( i )
        
        res = policy.random_search(max_num_probes=1, num_search_each_probe=batch_size, simulator=simulator())
        res = policy.bayes_search(max_num_probes=run_num-1, num_search_each_probe=batch_size, simulator=simulator(), score=score_mode, interval=1, num_rand_basis=1000)
        exec_n = res.total_num_search
        
        best_objs, best_acts = res.export_all_sequence_best_fx()
        objs, acts = res.fx[0:exec_n], res.chosen_actions[0:exec_n]
        
        df_obj = pd.DataFrame(objs, columns=['objs-'+str(i+1)])
        df_act = pd.DataFrame(acts, columns=['actn-'+str(i+1)])
        df_best_obj = pd.DataFrame(best_objs, columns=['best-objs-'+str(i+1)])
        df_best_act = pd.DataFrame(best_acts, columns=['best-acts-'+str(i+1)])
        
        df_objects = pd.concat([df_objects, df_obj], axis=1)
        df_actions = pd.concat([df_actions, df_act], axis=1)
        df_best_objs = pd.concat([df_best_objs, df_best_obj], axis=1)
        df_best_acts = pd.concat([df_best_acts, df_best_act], axis=1)
        
        res_all.append( best_objs )
    
    df_opt = pd.concat([df_objects, df_actions, df_best_objs, df_best_acts], axis=1)
    df_opt.to_csv(score_mode + '_batchsize_'+str(batch_size)+'_run_'+str(run_num)+'_optimization.csv')
    
    for x in res_all:
        ax.plot( x )
    
    plt.show()
```
